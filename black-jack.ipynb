# **Black Jack (Human vs Computer)**

```python
import random 

def shuffle_deck():
    ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'K', 'Q', 'J', 'A']
    suits = ['Spades', 'Hearts', 'Diamonds', 'Clubs']
    deck = [(suit, rank) for rank in ranks for suit in suits]
    random.shuffle(deck)
    return deck

def calculate_cards_in_hand_value(cards_in_hand):
    value = 0
    num_of_ace = 0
    
    for card in cards_in_hand:
        rank = card[1]
        if rank.isdigit():
            value += int(rank)
        elif rank in ['K', 'Q', 'J']:
            value += 10
        elif rank == 'A':
            value += 11
            num_of_ace += 1
    while value > 21 and num_of_ace:
        value -= 10
        num_of_ace -= 1
        
    return value

def print_cards_in_hand(cards_in_hand, card_to_be_hide=False):
    print("Cards in hand: ")
    for i, card in enumerate(cards_in_hand):
        if i == 0 and card_to_be_hide:
            print("Hidden Card")
        else:
            print(f"{card[1]} of {card[0]}")
    print("")

def blackjack():
    deck = shuffle_deck()
    player_cards = [deck.pop(), deck.pop()]
    computer_cards = [deck.pop(), deck.pop()]
    
    player_turn = True
    computer_turn = False
    
    # Player's loop
    while player_turn:
        print_cards_in_hand(player_cards)
        player_card_value = calculate_cards_in_hand_value(player_cards)
        print(f"Your hand value: {player_card_value}")
        
        if player_card_value > 21:
            print("Bust! You lose.")
            break
        elif player_card_value == 21:
            print("Blackjack! You win!")
            break
        next_choice = input("Do you want to 'stand' or 'hit'? ").lower()
        if next_choice == 'stand':
            player_turn = False
            computer_turn = True
        elif next_choice == 'hit':
            player_cards.append(deck.pop())
        else:
            print("Invalid input. Please enter 'hit' or 'stand'.")
            
    # Computer's loop
    while computer_turn:
        print_cards_in_hand(player_cards)
        print_cards_in_hand(computer_cards, card_to_be_hide=True)
        computer_card_value = calculate_cards_in_hand_value(computer_cards)

        while computer_card_value < 17:
            computer_cards.append(deck.pop())
            computer_card_value = calculate_cards_in_hand_value(computer_cards)

        print(f"Computer's hand value: {computer_card_value}")
        if computer_card_value > 21:
            print("Computer busted! You win.")
        elif computer_card_value == 21:
            print("Computer got Blackjack! You lose.")
        elif computer_card_value > player_card_value:
            print("Computer wins.")
        elif computer_card_value == player_card_value:
            print("Draw, Neither one is winner.")
        else:
            print("You win!")
            
        computer_turn = False

if __name__ == "__main__":
    blackjack()
```
