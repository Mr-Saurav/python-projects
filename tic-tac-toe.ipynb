# **1 Player Game (Human vs Computer)**

```python
import random

def sum(a, b, c):
    return a + b + c

def printBoard(xState, zState):
    for i in range(3):
        for j in range(3):
            index = i * 3 + j
            if xState[index] == 1:
                print('X', end=' ')
            elif zState[index] == 1:
                print('O', end=' ')
            else:
                print(index, end=' ')
            if j < 2:
                print('|', end=' ')
        print()
        if i < 2:
            print('--|---|--')

def checkWin(xState, zState):
    wins = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]
    for win in wins:
        if sum(xState[win[0]], xState[win[1]], xState[win[2]]) == 3:
            return 1  # X wins
        if sum(zState[win[0]], zState[win[1]], zState[win[2]]) == 3:
            return 0  # O wins
    return -1  # No winner yet

def computerMove(zState, xState):
    empty_cells = [i for i in range(9) if xState[i] == 0 and zState[i] == 0]
    return random.choice(empty_cells)

if __name__ == "__main__":
    xState = [0, 0, 0, 0, 0, 0, 0, 0, 0]
    zState = [0, 0, 0, 0, 0, 0, 0, 0, 0]
    print("Welcome to Tic Tac Toe")
    while True:
        printBoard(xState, zState)
        if sum(xState.count(1), zState.count(1), list(range(9)).count(1)) == 9: 
            #list(range(9)).count(1) it returns 1 always. so after playing 4 turns (each player) without winning, it return draw 
            print("\n It's a draw!")
            break
            
        if sum(xState.count(1), zState.count(1), list(range(9)).count(1)) % 2 == 0: 
            #list(range(9)).count(1) it returns 1 always, so we can switch the turn
            print("\n your Turn")
            value = int(input("Please enter a value: "))
            xState[value] = 1
        else:
            print("\n computer's Turn")
            value = computerMove(zState, xState)
            zState[value] = 1
            
        cwin = checkWin(xState, zState)
        if cwin != -1:
            printBoard(xState, zState)
            if cwin == 1:
                print("\n You Won the match")
            else:
                print("\n you lose the match")
            break
```

# **2 Players Game (Human VS Alien)**

```python
# Game Name: "HUMANS VS ALIENS"

# Here, "X" used by Humans and "O" used by Aliens

def galaxy(star, moon):

    # Display the Tic-Tac Toe board based on the current state of 'star' and 'moon' lists
    
    zero = 'X' if star[0] else ('O' if moon[0] else 0)
    one = 'X' if star[1] else ('O' if moon[1] else 1)
    two = 'X' if star[2] else ('O' if moon[2] else 2)
    three = 'X' if star[3] else ('O' if moon[3] else 3)
    four = 'X' if star[4] else ('O' if moon[4] else 4)
    five = 'X' if star[5] else ('O' if moon[5] else 5)
    six = 'X' if star[6] else ('O' if moon[6] else 6)
    seven = 'X' if star[7] else ('O' if moon[7] else 7)
    eight = 'X' if star[8] else ('O' if moon[8] else 8)
    
    print(f" {zero} | {one} | {two} ")
    print(f"---|---|---")
    print(f" {three} | {four} | {five} ")
    print(f"---|---|---")
    print(f" {six} | {seven} | {eight} ")

def sum(air, water, fire):

    # function to show Sum of the values
    
    life = air + water + fire
    return life

def winner(star, moon):

    # Check for a winner based on the Tic-Tac Toe board positions
    
    asteroid, meteor = 0,0
    win_situation = [[0,3,6], [1,4,7], [2,5,8], [0,1,2], [3,4,5], [6,7,8], [2,4,6], [0,4,8]]

    # Check for a winning condition
    
    for position in win_situation:
        if sum(star[position[0]], star[position[1]], star[position[2]]) == 3:
            print("Humans is the Winner of this game")
            return 1
        if sum(moon[position[0]], moon[position[1]], moon[position[2]]) == 3:
            print("Aliens is the Winner of this game")
            return 1

    # Check for a tie if all positions are filled
    
    for sky in range(0, 8):
        asteroid += star[sky]
        meteor += moon[sky]
    
    if asteroid == 5 or meteor == 5:
        print("Neither one is winner")
        return 2

    # No winner or tie
    return -1

if __name__ == "__main__":

    # Initialize Tic-Tac Toe board and player turns
    
    star = [0,0,0,0,0,0,0,0,0]
    moon = [0,0,0,0,0,0,0,0,0]
    turn = 1

    # Main game loop.
    
    print("\n\n-----------")
    print("--> Welcome to the Game Arena <--")
    print("-----------")
    print("--> Its Human vs Alien <--")
    print("--> Let's begin the game <-- \n")

    while True:
        galaxy(star, moon)

        try:
            if turn == 1:
                print("It's a Human's turn")
                value = int(input("choose & Enter place no: "))
                star[value] = 1
            else:
                print("It's an Alien's turn")
                value = int(input("choose & Enter place no: "))
                moon[value] = 1
        except ValueError:
            print("Invalid Input, Choose your empty place in between (0-8)")
            continue

        # Check for a winner or tie
        win = winner(star, moon)

        if win != -1:
            print("Match Over")
            break

        # Switch to turn the chance
        turn = 1 - turn
```
